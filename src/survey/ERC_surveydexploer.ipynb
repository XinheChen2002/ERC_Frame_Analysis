## Merge Advisors and Duration

```python
df = df.reset_index(drop=True)
if "respondent_id" not in df.columns:
    df["respondent_id"] = df.index
```

```python
advisor_cols = [c for c in df.columns if "advisor" in c.lower()]
advisor_cols
```

```python
connectors_base = "Based on your experience, which advisors acted as connectors or bridges—either between groups within this challenge or across different challenge years?"
types_base = "Which types of key advisors were most influential in shaping this challenge? Select up to three."
```

```python
connectors_cols = [c for c in df.columns if c.startswith(connectors_base)]
types_cols = [c for c in df.columns if c.startswith(types_base)]

connectors_cols, types_cols
```

```python
def make_year_map(base):
    return {
        base: 2021,
        base + ".1": 2022,
        base + ".2": 2023,
        base + ".3": 2024,
        base + ".4": 2025,
    }

connectors_year_map = make_year_map(connectors_base)
types_year_map = make_year_map(types_base)
```

```python
connectors_long = df[["respondent_id"] + connectors_cols].melt(
    id_vars="respondent_id",
    value_vars=connectors_cols,
    var_name="col",
    value_name="advisors_as_connectors"
)
connectors_long["year"] = connectors_long["col"].map(connectors_year_map)

connectors_long = (
    connectors_long
    .drop(columns="col")
    .dropna(subset=["advisors_as_connectors"])
    .reset_index(drop=True)
)

connectors_long.head(), connectors_long.shape
```

```python
connectors_long = connectors_long.merge(
    analysis_df_id[["respondent_id", "name"]],
    on="respondent_id",
    how="left"
)
connectors_long.head()
```
